```python
import pandas as pd, numpy as np
%matplotlib inline
import matplotlib
matplotlib.rc('figure', figsize=(8,8))
```

# From [Our world in data](https://ourworldindata.org/coronavirus-data)

## Italy

```python
world = pd.read_csv('daily-covid-cases-deaths.csv')
italy = world.loc[lambda df: df['Entity'] == 'Italy']
italy = italy.rename(columns={'Daily confirmed cases (cases)':'daily_cases',
                              'Daily confirmed deaths (deaths)':'daily_deaths'})
italy.insert(3, 'cases', np.cumsum(italy['daily_cases']))
italy.insert(3, 'deaths', np.cumsum(italy['daily_deaths']))
italy.plot.bar(y=['cases', 'daily_cases'], x='Date', subplots=True)            
```

```python
italy.plot.bar(y=['deaths', 'daily_deaths'], x='Date', subplots=True)
```
